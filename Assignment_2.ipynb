# **TASK #1**

```python
import pandas as pd
import numpy as np

# Data dictionary
data = {
    'Time': ['09:00', '11:00', '13:00', '15:00', '10:00', '12:00', '14:00', '16:00'],
    'Seats Available': [480, 480, 480, 640, 480, 480, 480, 640], 
    'Type': ['Departure', 'Departure', 'Departure', 'Departure', 'Arrival', 'Arrival', 'Arrival', 'Arrival' ],
    'Money Taken': [0,0,0,0,0,0,0,0],   # Each seat has a price of 25dollars
    'Original Seats': [480, 480, 480, 640, 480, 480, 480, 640]
}
# Create a DataFrame
df = pd.DataFrame(data)
df_departure = df[0:4]
df_arrival = df[4:8]
df_arrival.set_index(pd.Index(range(0, 4)), inplace=True)

def printdisp(df):
    for i in range (df.shape[0]):
        if df.iloc[i]['Seats Available'] == 0:
            df.at[i, 'Seats Available'] = 'CLOSED'
    print(df)
    for i in range (df.shape[0]):
        if df.iloc[i]['Seats Available'] == 'CLOSED':
            df.at[i, 'Seats Available'] = 0
```

# **TASK #2**

```python

price_per_ticket = 25 #Dollars 
tickets_price = 0


"""def entry_check(entered_item, df):
    status = False
    while True:
        for i in range (0, len(df)):
            if df.iloc['Time'][i] == entered_item:
                status = True
                break
        if status == True:
            return entered_item
        entered_item = input('You entered an invalid Time,' 
                            'Either it is not correct or the selected time has not enough seats.' 
                            'Please Try again')"""

def entry_check(entered_item, df):
    while True:
        if entered_item in df.values:
            return entered_item
        entered_item = input('You entered an invalid Time, '
                             'Either it is not correct or the selected time has not enough seats. '
                             'Please Try again: ')


def available_seats_check_1(df_departure, df_arrival, no_of_tickets):
    status1 = False
    status2 = False
    for i in range (0, len(df)):
        if (df_departure['Seats Available'].iloc[i]) >= no_of_tickets:
            status1 = True
            if (df_arrival['Seats Available'].iloc[i]) >= no_of_tickets:
                status2 = True
        
        if (status1 == True) and (status2 == True):
            return no_of_tickets
        else:       
            print('Seats are not available. Please Try another day')
            return 0
    
def available_seats_check_2(df_departure_arrival, selected_time, number_of_tickets):
    status = False
    while True:
        verified_time = entry_check(selected_time, df_departure_arrival['Time'])
        selected_time_list = df_departure_arrival[df_departure_arrival['Time'] == verified_time]
        selected_time_index = selected_time_list.index.tolist()[0]
        if (df_departure_arrival['Seats Available'].iloc[selected_time_index]) >= number_of_tickets:
            status = True
        if (status == True):
            return verified_time
        selected_time = input('Seats are not available in this time slot choose another time slot')
        
"""def fetch_index(df, verified_time):  
    selected_time = df[df['Time'] == verified_time]
    index = selected_time.index.tolist()[0]
    return index"""

def fetch_index(df, verified_time, type):
    selected_time = df[(df['Time'] == verified_time) & (df['Type'] == type)]
    index = selected_time.index.tolist()
    if not index:
        print(f"No matching index found for {verified_time} and {type}.")
        return None
    return index[0]

def discount(number_of_tickets):
    free_tickets = 0
    if number_of_tickets >=10:
        free_tickets = int(number_of_tickets/10)
        number_of_tickets -= free_tickets
    return free_tickets

def money_calculator(number_of_tickets):
    free_tickets = discount(number_of_tickets) 
    tickets_price = (number_of_tickets - free_tickets) * price_per_ticket * 2
    return free_tickets, tickets_price


def ticket_counter(df):

    free_tickets = 0
    number_of_tickets = int(input('How many tickets you want to purchase: '))
    number_of_tickets = available_seats_check_1(df_departure, df_arrival, number_of_tickets)
    
    printdisp(df_departure)
    ticket_time_depart = input('Choose the Departure time of your travel(e.g., 09:00): ')
    verified_time_depart = available_seats_check_2(df_departure, ticket_time_depart, number_of_tickets)    

    printdisp(df_arrival)
    ticket_time_arrive = input('Choose the Arrival time of your travel(e.g., 10:00): ')
    verified_time_arrive = available_seats_check_2(df_arrival, ticket_time_arrive, number_of_tickets)
    
    departure_index = fetch_index(df, verified_time_depart, 'Departure')
    arrival_index = fetch_index(df, verified_time_arrive, 'Arrival')
    
    free_tickets, tickets_price = money_calculator(number_of_tickets)
    return(tickets_price, number_of_tickets, free_tickets, price_per_ticket, departure_index, arrival_index)


def print_receipt(tickets_price, number_of_tickets, free_tickets, price_per_ticket):
    print('No. of tickets: ', number_of_tickets, '\n')
    print('Total Price: ', 2 * number_of_tickets*price_per_ticket, ' dollars\n') #2 for the return ticket
    print('Discount: ', free_tickets * price_per_ticket * 2, 'dollars\n')
    print('Price to be paid: ', tickets_price, ' dollars\n')

def updated_screen_display(df, number_of_tickets, departure_index, arrival_index):
    df.at[departure_index, 'Seats Available'] -= number_of_tickets
    df.at[arrival_index, 'Seats Available'] -= number_of_tickets
    df.at[departure_index, 'Money Taken'] += tickets_price/2
    df.at[arrival_index, 'Money Taken'] += tickets_price/2
    df_departure = df[0:4]
    df_arrival = df[4:8]
    df_arrival.set_index(pd.Index(range(0, 4)), inplace=True)
    return df, df_departure, df_arrival




```

# **Code Runner**

```python
tickets_price, number_of_tickets, free_tickets, price_per_ticket, departure_index, arrival_index = ticket_counter(df)
df, df_departure, df_arrival = updated_screen_display(df, number_of_tickets, departure_index, arrival_index)
print_receipt(tickets_price, number_of_tickets, free_tickets, price_per_ticket)
```

```python
def end_of_day(df):
    total_passengers = 0
    for i in range(0, len(df)):
        passenger_for_this_train = df['Original Seats'].iloc[i] - df['Seats Available'].iloc[i]
        total_passengers += passenger_for_this_train
        if df['Seats Available'].iloc[i] < df['Seats Available'].iloc[i-1]:
            busiest_train_index = i
    total_money_taken = df['Money Taken'].sum()
    
    print('Total Passengers for today: ', total_passengers, '\n'
          'Total Money Taken: ', total_money_taken, '\n'
          'Busiest Train: ', df['Time'].iloc[busiest_train_index])
    

end_of_day(df)
```
