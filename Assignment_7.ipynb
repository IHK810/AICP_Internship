```python
# Constants
NUM_CHARITIES = 3

# Data structure to store charity information
charities = {i: {'name': input(f"Enter the name of Charity {i}: "), 'total_donation': 0} for i in range(1, NUM_CHARITIES + 1)}

# Function to display charity names and numbers
def display_charities():
    print("\nCharities:")
    for i, charity_info in charities.items():
        print(f"{i}. {charity_info['name']}")

# Function to calculate donation
def calculate_donation(bill_amount, charity_choice):
    return 0.01 * bill_amount

# Function to record and total each donation
def record_and_total_donation():
    display_charities()

    # Input charity choice
    while True:
        charity_choice = int(input("\nEnter the number of the charity (1, 2, or 3): "))
        if 1 <= charity_choice <= NUM_CHARITIES:
            break
        else:
            print("Invalid choice. Please choose 1, 2, or 3.")

    # Input bill amount
    bill_amount = float(input("Enter the value of the customer's shopping bill: "))

    # Calculate donation
    donation = calculate_donation(bill_amount, charity_choice)

    # Update total donation for the chosen charity
    charities[charity_choice]['total_donation'] += donation

    # Output the name of the charity and the amount donated
    print(f"\nDonation to {charities[charity_choice]['name']}: ${donation:.2f}")

# Function to show the totals so far
def show_totals_so_far():
    grand_total = 0

    # Sort charities by total donation in descending order
    sorted_charities = sorted(charities.items(), key=lambda x: x[1]['total_donation'], reverse=True)

    print("\nCharity Totals So Far:")
    for charity_number, charity_info in sorted_charities:
        total_donation = charity_info['total_donation']
        grand_total += total_donation
        print(f"{charity_info['name']}: ${total_donation:.2f}")

    # Output grand total
    print(f"\nGRAND TOTAL DONATED TO CHARITY: ${grand_total:.2f}")

# Test the functions
# Task 1
display_charities()

# Task 2
record_and_total_donation()
record_and_total_donation()

# Task 3
record_and_total_donation()
show_totals_so_far()
```
